# Exploratory Data Analysis (EDA) on Retail Sales Data
## Internship Project
This project aims to analyze retail sales data to uncover patterns, trends, and insights that can help the business make informed decisions.

## 1. Data Loading & Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv('DB1 store_cities.csv')
df2 = pd.read_csv('DB1 sales.csv')
df3 = pd.read_csv('DB1 product_hierarchy.csv')

# Display first few rows
df2.head()
```

## 2. Descriptive Statistics

```python
# Summary statistics for sales and revenue
df2[['sales', 'revenue', 'stock', 'price']].describe()
```

## 3. Time Series Analysis
Analyzing sales trends over time.

```python
# Convert date column to datetime
df2['date'] = pd.to_datetime(df2['date'])

# Aggregate sales by date
sales_trend = df2.groupby('date')['sales'].sum()

# Plot sales trend
plt.figure(figsize=(12,6))
sales_trend.plot()
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Time')
plt.show()
```

## 4. Product & Customer Analysis

```python
# Top selling products
top_products = df2.groupby('product_id')['sales'].sum().sort_values(ascending=False)[:10]

# Plot top products
plt.figure(figsize=(12,6))
sns.barplot(x=top_products.index, y=top_products.values)
plt.xlabel('Product ID')
plt.ylabel('Total Sales')
plt.title('Top Selling Products')
plt.xticks(rotation=45)
plt.show()
```

## 5. Visualization
Using bar charts, line plots, and heatmaps to analyze data.

```python
# Heatmap of sales correlation with other features
plt.figure(figsize=(10,6))
sns.heatmap(df2[['sales', 'revenue', 'stock', 'price']].corr(), annot=True, cmap='coolwarm')
plt.title('Sales Correlation Heatmap')
plt.show()
```
